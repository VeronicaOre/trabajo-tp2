```python
import xlrd
import requests
import re
```

```python
path = "C:\\Users\\pablo\\Downloads\\genes.xlsx"

el_excel = xlrd.open_workbook(path)
hoja_excel = el_excel.sheet_by_index(0)

"""
print('tu hoja tiene '+ str(hoja_excel.ncols)+' columnas')
print('tu hoja tiene '+ str(hoja_excel.nrows)+' filas')
"""

scrapeo = []
for row in range(hoja_excel.nrows):
    if row!=0:
        dato=hoja_excel.cell_value(row,13)
    else:
        continue
    gen = re.findall('^([A-Z0-9]+)',dato); gen=str(gen[0])
    url = 'https://www.ncbi.nlm.nih.gov/gene/?term='+gen+'+bos+taurus'
    r=requests.get(url)
    informacion=r.text[:]
    Chromosome=re.findall('NC.[0-9.]+..([0-9.]+)',informacion)
    print(Chromosome[0])
    scrapeo.append(Chromosome[0])
    
```

```python
import pandas as pd
from openpyxl import load_workbook
import pandas.io.formats.excel
```

```python
df  = pd.DataFrame({'Rango': scrapeo})
```

```python
book = load_workbook('genes.xlsx')
writer = pd.ExcelWriter('genes.xlsx', engine='openpyxl') 
writer.book = book
```

```python
pandas.io.formats.excel.header_style = None
```

```python
writer.sheets = dict((ws.title, ws) for ws in book.worksheets)
df.to_excel(writer, book.worksheets[0].title, startcol = 14,  index = False) 
writer.save()
```
